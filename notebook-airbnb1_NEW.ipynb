# **Can you spot any major differences in the Airbnb market between cities?**

```python
#Sorting the group in descending order
sorted_df1 = df1.groupby('city')['instant_bookable'].count().sort_values(ascending = False)
print(sorted_df1)
```

```python
sorted_df1.head(30).to_frame().plot(kind='bar',figsize=(15,8))
for i, v in enumerate(sorted_df1):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.ylabel('City')
plt.title('Instant_Bookable')
```

# **Which attributes have the biggest influence on price?**

```python
#Grouping bedrooms with city & also sort by ascending
x = df1.groupby("city")['bedrooms'].count().sort_values(ascending = False)
x
```

```python
x.head(30).to_frame().plot(kind='bar',figsize=(15,8))
for i, v in enumerate(x):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.ylabel('City')
plt.title('Bedrooms')
```

```python
#Grouping review_scores_rating with city
y = df1.groupby("city")['review_scores_rating'].mean().sort_values(ascending = False).round(2)
y
```

```python
y.head(30).to_frame().plot(kind='bar',figsize=(15,8))
for i, v in enumerate(y):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.ylabel('City')
plt.title('average_review_scores_rating')
```

```python
#Grouping review_scores_cleanliness with city
z= df1.groupby("city")['review_scores_cleanliness'].mean().sort_values(ascending = False).round(2)
z
```

```python
z.head(30).to_frame().plot(kind='bar',figsize=(15,8))
for i, v in enumerate(z):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.ylabel('City')
plt.title('average_review_scores_cleanliness')
```

```python
#Grouping price with city
a= df1.groupby("city")['price'].mean().sort_values(ascending = False).round(2)
a
```

```python
a.head(30).to_frame().plot(kind='bar',figsize=(15,8))
for i, v in enumerate(a):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.ylabel('City')
plt.title('average_price')
```

# **Are you able to identify any trends or seasonality in the review data?**

```python
#Reading the Dataframe of Reviews.csv 
df2= pd.read_csv('/kaggle/input/airbnb-listings-reviews/Airbnb Data/Reviews.csv')
df2.head()
```

```python
df2.tail(5)
```

```python
df2.shape
```

```python
df2.info()
```

```python
df2.describe()
```

```python
#Merging both lf and rf into one Dataframe
merge_df = df1.merge(df2, how='inner', on=['listing_id'])
merge_df.head()
```

```python
merge_df.info()
```

```python
merge_df.shape
```

```python
#Creating a new columns only of Months
merge_df['Months'] = pd.to_datetime(merge_df['date'], infer_datetime_format=True).dt.month
merge_df.info()
```

```python
#Getting the sorted values of date with decending order
merge_df['date'].value_counts().sort_values(ascending=False)
```

```python
#Display the Graph
merge_df.groupby('Months').agg('Months').count().to_frame().plot(kind='bar',figsize=(15,8))


plt.ylabel('No.reviews')
plt.title('reviews per Month')
```

```python
merge_df.groupby('Months').agg('Months').count()
```

```python
merge_df.groupby('Months').agg('Months').count().max()
```

```python
merge_df.groupby('Months').agg('Months').count().min()
```

* In dataset overall all months perform 400000+ reviews but in 4 months their is difference.
* month 4&5 their low reviews like 375740 & 398185 resp.
* month 9&10 their high reviews like 511245 & 552877 resp.
